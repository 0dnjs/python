```python
# 크롬 드라이버
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By

import time

driver = webdriver.Chrome()

driver.get('https://www.naver.com/')

time.sleep(3)

searchValue = input("검색어 입력: ")

searchInput = driver.find_element(by=By.CSS_SELECTOR, value="#query")
searchInput.send_keys(searchValue) # 입력
searchInput.send_keys(Keys.RETURN) # 엔터

time.sleep(2)

# 오른쪽으로 이동버튼 클릭
driver.find_element(by=By.XPATH, value='//*[@id="lnb"]/div[1]/div/div[1]/div/div[2]/div[2]/a').click()
time.sleep(1)

#뉴스탭 클릭
driver.find_element(by=By.XPATH, value='//*[@id="lnb"]/div[1]/div/div[1]/div/div[1]/div[7]/a').click()
time.sleep(2)

# 뉴스타이틀 크롤링
newsTitles = driver.find_elements(by=By.CSS_SELECTOR, value='.news_tit')
for title in newsTitles:
    print(title.text)

for i in range(2, 5):
    driver.find_element(by=By.XPATH, value=f'//*[@id="main_pack"]/div[2]/div/div/a[{i}]').click()
    time.sleep(2)
    newsTitles = driver.find_elements(by=By.CSS_SELECTOR, value='.news_tit')
    for title in newsTitles:
        print(title.get_attribute(name="href"))
    newsImages = driver.find_elements(by=By.CSS_SELECTOR, value=".dsc_thumb .thumb")
    for img in newsImages:
        print(img.get_attribute(name="src"))

# //*[@id="wrap"]/header/div[3]/nav/ul/li/[2]/a
# //*[@id="wrap"]/header/div[3]/nav/ul/li/[3]/a 웹페이지에서 ctrl + shift + c 눌러서 버튼클릭 후 붙여넣으면 (마지막 3이 들어간 자리) 페이지버튼의 경로가 나옴 

webToonList = [
    {
        "dayofTheWeek": "월",
        "webtoon": [
            {
                "title": "참교육",
                "author": "채용택 / 한가람",
                "grade": "9.87",
                "postImage": "https://asdfsdafsdf.jpg"
            },
            {
                "title": "퀘스트지상주의",
                "author": "박태준 만화회사",
                "grade": "9.75",
                "postImage": "https://asdfsdafsdf.jpg"
            }
        ]
    }
]
```
